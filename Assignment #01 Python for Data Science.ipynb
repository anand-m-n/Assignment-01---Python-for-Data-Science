#### Visualization 

```python
sns.countplot(x="Classification", data=iris_data, hue='Classification')
plt.legend(bbox_to_anchor=(1.01, 1),
           borderaxespad=0)
plt.title("Count plot for the species")
```

There are 50 records for each Iris species

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='SL',y='SW', data=iris_data, hue= 'Classification')
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Comparision between species based on sepal length and width")
```

Iris Setosa has sepals with smaller lenght but they have the higher width.
Iris Virginica has sepals with larger length but smaller width

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='PL',y='PW', data=iris_data, hue= 'Classification')
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("Comparision between species based on petal length and width")
```

Iris setosa has petals with smaller length as well as smaller width. 
Iris Virginica has petals with larger length and width

```python
fig, axes = plt.subplots(2,2,figsize=(15,8))
sns.boxplot(x='Classification', y= 'PL', data=iris_data, ax = axes[0,0])

sns.boxplot(x='Classification', y= 'SL', data=iris_data,ax = axes[0,1])

sns.boxplot(x='Classification', y= 'PW', data=iris_data, ax = axes[1,0])

sns.boxplot(x='Classification', y= 'SW', data=iris_data, ax = axes[1,1])
plt.suptitle('Box plots of Iris Species for different features')

```

Iris Setosa has the lowest sepal and petal lenghths, but they have the highest sepal width


```python
sns.heatmap(iris_data.corr(), annot = True, cmap='Greens')
plt.title("Heat Map")
```

Petal Length is highly correlated (positively) with Petal width and sepal length.
Sepal length shows neagative correlation with all other features.

```python
fig, axes = plt.subplots(2,2,figsize=(15,8))

sns.histplot(x = 'PL', data=iris_data, hue='Classification', kde =True, ax=axes[0,0])
sns.histplot(x = 'SL', data=iris_data, hue='Classification', kde = True, ax=axes[1,0])
sns.histplot(x = 'PW', data=iris_data, hue='Classification', kde = True, ax=axes[0,1])
sns.histplot(x = 'SW', data=iris_data, hue='Classification', kde = True, ax=axes[1,1])
```

```python
import warnings
warnings.filterwarnings('ignore')
sp = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
fig, axes = plt.subplots(2,2,figsize=(15,8))
for species in sp:
    subset = iris_data[iris_data['Classification'] == species]
    s1 =sns.distplot(subset['PL'], hist = True, kde = True, ax = axes[0,0])
    s2=sns.distplot(subset['PW'], hist = True, kde = True, ax = axes[0,1])
    s3=sns.distplot(subset['SL'], hist = True, kde = True, ax = axes[1,0])
    s4=sns.distplot(subset['SW'], hist = True, kde = True, ax = axes[1,1])
plt.suptitle('Density plots of Iris Species for different features')

fig.legend([s1, s2, s3, s4],     
           labels=sp,  
           loc="center right",   
           borderaxespad=0.1,    
           title="Species" 
           )

```

Iris setosa is clearly distinguishable from other species by its smaller petal features.
Versicolor and virginica species have some overlapping petal features.
In general, petal features can be used to distinguish between the species in a better way.
